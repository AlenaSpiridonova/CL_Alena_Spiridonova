1. (re.findall) Напишите регулярное выражение, которое вытаскивало бы все адреса из документа addresses.txt

```python
import re


input_text = open('Files\\addresses.txt', 'r').read()
addresses = re.findall(r'^\d.*?\d{5}', input_text, flags=re.MULTILINE | re.DOTALL)
print('\n\n'.join(addresses))
```
